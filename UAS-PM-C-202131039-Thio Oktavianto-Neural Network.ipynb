```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv('harga_rumahh.csv')
dataset
```

```python
# data preprocessing

# cek jika ada data null atau NaN
dataset.isnull().sum()
```

```python
# cek data terduplikat
dataset.duplicated().sum()
```

```python
# drop data duplikat
dataset = dataset.drop_duplicates()
print(dataset.duplicated().sum())
```

```python
x = np.asarray(dataset.drop(['price'], axis = 1), dtype=np.int32)
y = np.asarray(dataset['price'], dtype=np.int32)
```

```python
x.shape
```

```python
import collections
collections.Counter(y)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, 
                                                    y, 
                                                    test_size=0.3, 
                                                    random_state=0)
```

```python
from sklearn.neural_network import MLPClassifier

model = MLPClassifier(hidden_layer_sizes=(150,1), 
                      max_iter=450, 
                      random_state=1)

model.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score

y_pred = model.predict(x_train)
print(accuracy_score(y_train, y_pred))

y_pred_test = model.predict(x_test)
print(accuracy_score(y_test, y_pred_test))
```

```python
man1 = np.array([[1340, 1.50, 3.0, 1.5]])
print(model.predict(man1))
```

```python
from sklearn.model_selection import GridSearchCV

params = {
    'hidden_layer_sizes' : [(150,100,50)],
    'max_iter' : [500],
    'activation' : ['identity', 'logistic', 'tanh', 'relu'],
    'solver' : ['lbfgs', 'sgd', 'adam'],
    'learning_rate' : ['constant', 'invscaling', 'adaptive'],
    'random_state' : [1]
}

grid_search = GridSearchCV(estimator=MLPClassifier(),
                          param_grid = params,
                          n_jobs = 3,
                          verbose = 1,
                          scoring = 'accuracy')

grid_search.fit(x_train, y_train)
```

```python
print(grid_search.best_score_)
```

```python
print(grid_search.best_estimator_.get_params())
```
